```python
noisy_set_sizes0 = {
    'mislabeled' : (7600, None),                   # max size: 15000
    # 'irrelevant' : (8600, [0.25,0.25,0.25,0.25]),  # max size: 34259
    # 'translated' : (8600, "reserve_labels"),       # max size: 5000
}

noisy_set_sizes1 = {
    # 'mislabeled' : (8600, None),                   # max size: 15000
    'irrelevant' : (8600, [0.25,0.25,0.25,0.25]),  # max size: 34259
    # 'translated' : (8600, "reserve_labels"),       # max size: 5000
}

noisy_set_sizes2 = {
    # 'mislabeled' : (8600, None),                   # max size: 15000
    # 'irrelevant' : (8600, [0.25,0.25,0.25,0.25]),  # max size: 34259
    'translated' : (8600, 0),       # max size: 5000
}

noisy_set_sizes3 = {
    # 'mislabeled' : (8600, None),                   # max size: 15000
    # 'irrelevant' : (8600, [0.25,0.25,0.25,0.25]),  # max size: 34259
    'translated' : (8600, 0.50),                   # max size: 5000
}

noisy_set_sizes4 = {
    # 'mislabeled' : (8600, None),                   # max size: 15000
    # 'irrelevant' : (8600, [0.25,0.25,0.25,0.25]),  # max size: 34259
    'translated' : (8600, 1),                   # max size: 5000
}
# The training set of each experiment
origin_train_set_sizes_part1 = [2000, 4000, 5000, 8000]
origin_train_set_sizes_part2 = [ 10000, 15000, 20000]
noisy_train_set_sizes  = [(4000, 1000), (8000, 2000), (12000,3000), (15000, 5000)]

NoisySetExperimentTitles = ['mislabeled(100%)', 'irrevalent(100%)', 'translated(100%) 0% label changed', 'translated(100%) 50% label changed', 'translated(100%) 100% label changed']
NoisyDataSizeTitle = ['(4000,1000)', '(8000,2000)', '(12000,3000)', '(15000,5000)']

def GetNoisyExperimentFilePath( ExperimentIndex, DataSizeIndex):
    path = 'saving/Hist_NoisySetSize' + str(ExperimentIndex) + "_" + str(DataSizeIndex) + ".csv"
    return path
    
def BuildNoisyExperimentTitle(ExperimentIndex, DataSizeIndex):
    expTitle = NoisySetExperimentTitles[ExperimentIndex]
    sizeTitle = NoisyDataSizeTitle[DataSizeIndex]
    title = "Noisy Set:" + sizeTitle + " " + expTitle
    return title
    
def DisplayNoisyHistory(ExperimentIndex, DataSizeIndex):
    filepath = GetNoisyExperimentFilePath(ExperimentIndex, DataSizeIndex)
    df = pd.read_csv(filepath)
    title = BuildNoisyExperimentTitle(ExperimentIndex,DataSizeIndex)
    BERTModel.PrintLossWithHistory(title, df)
    BERTModel.PrintAccuracyWithHistory(title, df)
    
def GetOriginTrainExperimentFilePath(part, index):
    filePath = 'saving/Hist_OriginTrainDataPart' + str(part+1) + "_" + str(index) + ".csv"
    return filePath

OriginTrainSizeTitles = [["2000", "4000", "5000", "8000"], ["10000", "15000", "20000"]]

def BuildOriginTrainExperimentTitle(part, index):
    sizeStr = OriginTrainSizeTitles[part][index]
    title = "Train Set " + sizeStr
    return title

def DisplayOrigDataTrainHistory(part, index):
    filepath = GetOriginTrainExperimentFilePath(part, index)
    title = BuildOriginTrainExperimentTitle(part, index)
    df = pd.read_csv(filepath)
    BERTModel.PrintLossWithHistory(title, df)
    BERTModel.PrintAccuracyWithHistory(title, df)
    pass
```
